# GAN model

```python
class GAN_pretrained_gen:
    """Conditional Generative Adversarial model based on pretrained model for generator"""
    
    # -------------------------------------------------------------------------------------------------------
    #                INITIALIZATION
    # ------------------------------------------------------------------------------------------------------
    
    def __init__(self, 
                 path_gen, 
                 output_dir,
                 input_shape=(256,256,128,1), 
                 lr=0.0002, 
                 epoch=100,
                 beta=0.9,
                 disc_loss = 'BinaryCrossentropy',
                 gan_loss = 'BinaryCrossentropy',
                 seed=123456,
                last_layer_encoder='activation_14',
                filters_disc=1,
                dense_neuron_disc=5):
        
        
        # define path pretrained model for generator
        self.path_gen=path_gen
        self.output_dir=output_dir
        
        # Input sizes
        self.img_size = input_shape
        
        # compiler/model info
        self.lr = lr
        self.beta=beta
        self.epoch=epoch
        self.seed=seed
        self.disc_loss=disc_loss
        self.gan_loss=gan_loss
        self.last_layer_encoder=last_layer_encoder
        self.filters_disc=filters_disc
        self.dense_neuron_disc=dense_neuron_disc
        
        
            
        # bias and weights initialization (for disc model) --------------------------------------------
        self.kernel_initializer  = tf.keras.initializers.GlorotNormal(seed=seed)
        self.bias_initializer =tf.keras.initializers.Zeros()    
        
        # build models
        self.create_gen_model()
        self.create_disc_model()
        self.create_GAN_model()
        
        # save best model
        #model_checkpoint_callback = ModelCheckpoint(
        #filepath='saved_model',
        #save_weights_only=False,
        #monitor=g_loss,
        #mode='min',
        #save_best_only=True) 
        
        # initialize FID metric
        self.fid_gan=None
              
        
    # -------------------------------------------------------------------------------------------------------
    #                GENERATOR
    # ------------------------------------------------------------------------------------------------------    
    # Note: for now the seed is not used ??? have to check how to do that for an already existng model? 
    
    def create_gen_model(self):
        
        # fetch the Generator
        self.gen = load_model(self.path_gen)
        self.gen._name='gen'
        
        
     
    # -------------------------------------------------------------------------------------------------------
    #                DISCRIMINATOR
    # ------------------------------------------------------------------------------------------------------
    # idea: what about freezing the pretrained discriminator weight (from the trained model) 
    # and only trained the last layers? (not implemented here)
    
    def create_disc_model(self):
    
        # extraction of encoder 
        model = Model(inputs = self.gen.layers[0].input, 
                      outputs = self.gen.get_layer(self.last_layer_encoder).output)

        # add layers ----------------------------------------------------------
        x=Conv3D(self.filters_disc, (1, 1, 1), padding="same", 
                                   activation='relu',
                                   kernel_initializer=self.kernel_initializer,
                                   bias_initializer=self.bias_initializer,name='Conv3D_kernel_1')(model.output)
        x=Flatten()(x)
        #x=Dropout(0.5,name='dropout_init_new')(x)
        x=Dense(self.dense_neuron_disc,activation='relu',
                kernel_initializer=self.kernel_initializer,
                    bias_initializer=self.bias_initializer,
                name='Dense_1')(x)
       
        x=Dense(1,activation='sigmoid')(x)
        
        # define disc model -------------------------------------------------------
        self.disc = Model(inputs = self.gen.layers[0].input, 
                           outputs = x, 
                           name='disc')
    
        # compile the model -------------------------------------------------------
        self.disc_opt=Adam(learning_rate=self.lr, 
                      beta_1=self.beta,
                      decay = self.lr )
        
        self.disc.compile(loss=self.disc_loss, 
                           optimizer=self.disc_opt,
                           metrics=['accuracy'])
           
    
    # -------------------------------------------------------------------------------------------------------
    #                GAN
    # ------------------------------------------------------------------------------------------------------
    # still not sure if the discriminator weights must be set to trainabel after the GAN is compiled
    
    def create_GAN_model(self):
        
        # freeze the disc weights
        self.disc.trainable=False
        
        # define input/output
        gan_input=Input(self.img_size,name='input')
        fake_image=self.gen(gan_input)
        gan_output=self.disc(fake_image)
        
        # define model
        self.gan=Model(inputs=gan_input,
                       outputs=gan_output,name='GAN') 
        
        # define optimizer
        self.gan_opt = Adam(learning_rate=self.lr/2, 
                       beta_1=0.9, 
                       decay=self.lr/ self.epoch)
        
        # compile the model
        self.gan.compile(loss=self.gan_loss,
                         optimizer=self.gan_opt,
                         metrics=['accuracy'])   
        #self.disc.trainable=True
    
    
    # ------------------------------------------------------------------------------------------
    #                    Train
    # --------------------------------------------------------------------------------------------
        
    def train(self, 
              X_train,
              Y_train,
              X_val=None,
              Y_val=None,
              batch_size=3, 
              cpkt_interval=50,
             gen_interval=10):
        
        
    
        tf.keras.backend.clear_session()
        data_train_batch = tf.data.Dataset.from_tensor_slices((X_train, Y_train)).batch(batch_size)
        #data_test_batch=tf.data.Dataset.from_tensor_slices((X_test, Y_test)).batch(batch_size)
        
        # for save generator -----------------
        criteria=9999
        
        # --------------------------------------
        
        
        for epochs in range(self.epoch):
            
            # test to use checkpoints
            #ckpt_manager = self.use_checkpoints()
            
            # ------------------------
            
            for (X,y) in data_train_batch:
        
            
                batch_size=X.shape[0]
                # generate images ------------------------------------------
                fake_x = self.gen.predict(X)       
                real_x=y
                x=np.concatenate((real_x,fake_x))
                
            
                # discriminator -----------------------------------
                # half of the disc y to 0.9 to correlate with x       
                disc_y = np.zeros(2*batch_size)
                disc_y[:batch_size] = 1#0.9
        
                # train discriminator -----------------------------
                #disc.trainable=True
                d_loss = self.disc.train_on_batch(x,disc_y)
            
                # train generator -----------------------------------------
                y_gen = np.ones(batch_size)
                g_loss = self.gan.train_on_batch(X,y_gen)
                
        
            print('g_loss = ' + str(g_loss[0]))
            print('epoch ' + str(epochs) + '/' + str(self.epoch) )
            # save best model

            
            # evaluate models ----------------------------------------------
            #val_gen_loss,val_disc_loss = self.evaluate_models(X_val, Y_val, criteria=criteria)
            #gan_fid=self.evaluate_models(X,Y, criteria=criteria)
            #criteria = gan_fid
            # save things --------------------------------------------------
            wandb.log({
                       "disc_loss_train":d_loss[0],
                       "disc_accuracy_train":d_loss[1],
                       "gan_loss_train":g_loss[0] ,
                        "gan_accuracy_train":g_loss[1]
                       # "gan_FID":gan_fid
                        #"disc_loss_val" : val_disc_loss[0],
                        #"disc_accuracy_val" : val_disc_loss[1],
                        #"gen_mse_val" : val_gen_loss
                      })
            model_checkpoint_callback = ModelCheckpoint(
                            filepath='saved_model/Best_model',
                            save_weights_only=False,
                            monitor=g_loss[0],
                            mode='min',
                            save_best_only=True)     
  
        
        
            # checkpoint management --------------------------------------------------------
            if (epochs + 1) % cpkt_interval == 0:
                
                ckpt_manager = self.use_checkpoints()
                ckpt_save_path = ckpt_manager.save()
                print ('Saving checkpoint for epoch {} at {}'.format(epochs+1,ckpt_save_path))
            
        
            if (epochs + 1) % gen_interval == 0:
                print('saving gen model')
                location='gen/gen_epoch_'+str(epochs)
                self.gen.save(self.output_dir+location)
                plot_data(self.gen,
                            X_train,
                            model_dir=self.output_dir+location)
                
                location='disc/disc_epoch_'+str(epochs)
                self.disc.save(self.output_dir+location)
                
                # calculate FID -- does not work because it requires a 2D model
                #self.gan_fid=calculate_FID(self.gen,Y_train,X_train)
                
                
                
            

    
        
    
    
    
    # -------------------------------------------------------------------------------
    #                plot results
    # ---------------------------------------------------------------------------------
    
    # TO DO : possibly upload fig to W&B
    
    # -------------------------------------------------------------------------------
    #                Save  model
    # ---------------------------------------------------------------------------------
    
    # TO DO : find best criteria (mse gen ?)
    
    def save_optimizer_weights(self,model,model_name):
        #for now not used --------------------------------------
        # save optimizer weights ------------------------------
        #model.save_weights('weights.h5')
        symbolic_weights = getattr(model.optimizer, 'weights')
        weight_values = tf.keras.backend.batch_get_value(symbolic_weights)
        with open(self.save_path_model+'/optimizer_'+ model_name+'.pkl', 'wb') as f:
                pickle.dump(weight_values, f)
        # how to load the optimizer weights:
        #model._make_train_function()
        # How to set weights to optimizer model
        #with open('optimizer.pkl', 'rb') as f:
        #    weight_values = pickle.load(f)
        #model.optimizer.set_weights(weight_values)
        
    def save_models(self,model):
        # for now not used ------------------------------
        discriminator.trainable = False
        save_model(gan, 'gan')
        discriminator.trainable = True
        save_model(generator, 'generator')
        save_model(discriminator, 'discriminator')
        
        
        
    def use_checkpoints(self):
        checkpoint_dir = self.output_dir+'/checkpoints'
        checkpoint = tf.train.Checkpoint(     gan_opt=self.gan_opt,
                                              disc_opt=self.disc_opt,
                                              gen=self.gen,
                                              disc=self.disc,
                                              gan = self.gan )
          
        # what does max_to_keep do?
        ckpt_manager = tf.train.CheckpointManager(checkpoint, 
                                                  checkpoint_dir, 
                                                  max_to_keep=50)
    
        return ckpt_manager
        
        #if ckpt_manager.latest_checkpoint:
        #    checkpoint.restore(ckpt_manager.latest_checkpoint)  
        #    print ('Latest checkpoint restored!!')

        
    # -------------------------------------------------------------------------------
    #                Evaluate model
    # ---------------------------------------------------------------------------------
    
# previous eveluation based on validation ----------------------------------------------------------------
#     def evaluate_models(self,X_val,Y_val, criteria = 9999):
        
#         fake_x = self.gen.predict(X_val)       
#         real_x=Y_val
        
#         val_gen_loss=self.evaluate_gen(fake_x, Y_val)
#         val_disc_loss = self.evaluate_disc(real_x,fake_x)
        
#         if val_gen_loss<criteria:
#             self.gen.save(self.output_dir+'gen/best_gen_model')
        
#         return val_gen_loss,val_disc_loss
        
    
    
#     def evaluate_disc(self,real_x,fake_x):
               
#         x=np.concatenate((real_x,fake_x))
        
#         disc_pred=self.disc.predict(x)
        
#         batch_size=len(real_x)    
#         # discriminator -----------------------------------
#         # half of the disc y to 0.9 to correlate with x       
#         disc_y = np.zeros(2*batch_size)
#         disc_y[:batch_size] = 1#0.9
        
        
#         val_disc_loss= self.disc.evaluate(disc_pred,disc_y) 
        
#         return val_disc_loss
    
    
#     def evaluate_gen(self,fake_x,real_x):
        
#         val_gen_loss = np.sum(np.sqrt(np.square(np.squeeze(fake_x-real_x))))/len(fake_x)#mean_squared_error(fake_x, real_x)
        
#         return val_gen_loss
        
#---------------------------------------------------------------------------------------------  
# Evalutation based on FID ----------------------------------------------------------------
        
    def evaluate_GAN(self, X, Y):
        # evaluate GAN model based on frechet Inception distance
        fd = fid.FrechetInceptionDistance(self.gen, (0,1)) 
        gan_fid = fd(Y, X)
        return gan_fid
    
    
    def evaluate_models(self,X,Y, criteria = 9999):
        
        
        gan_fid=evaluate_GAN(X,Y)
        
        if gan_fid<criteria:
            self.gen.save(self.output_dir+'gen/best_gen_model')
        
        return gan_fid
        
    
    
    
    
    
```
