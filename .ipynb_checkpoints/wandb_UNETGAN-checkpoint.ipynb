# Model development

```python
class create_generator_discriminator:
    
    def __init__(self, X_train,
                 model,
                 start_neurons = 32,
                 size_layers=[2,4,8,16],
                 number_conv = 1,
                 dropout_regularization=False,
                 stride=(1,1,1),
                 dropout=0.5,
                 kernel_size=(3,3,3),
                 residual_connection=False,
                 maxpool=(1,1,1),
                    seed=None):
        
        
        
        self.start_neurons = start_neurons
        self.size_input = X_train.shape
        self.size_layers=size_layers
        self.number_conv = number_conv
        self.dropout_regularization=dropout_regularization
        self.stride=stride
        self.dropout=dropout
        self.kernel_size=kernel_size
        self.residual_connection=residual_connection
        self.input_layer = Input(np.shape(X_train)[1:])
        self.maxpool = maxpool
        self.kernel_initializer = tf.keras.initializers.GlorotNormal(seed=seed)
        self.bias_initializer = tf.keras.initializers.Zeros()
        
        if model=='generator':
            self.gen_model()
            self.model_name='generator'
        elif model=='discriminator':
            self.disc_model()
            self.model_name='discriminator'
        
        
    

    def gen_model(self):

        
        sizes_down=np.flip(self.size_layers)
    
        # for data size consistency ----------------------------------------
        conv_good_size = ZeroPadding3D(padding=((0, 0), (0, 0), (0, 128-self.size_input[3])))(self.input_layer)
        # --------------------------------------------
       
        
        # encoder ----------------------   
        conv1=self.UNet_conv_layers(start_neurons=self.start_neurons,
                         kernel_size=self.kernel_size,
                         maxpool=self.maxpool,
                         dropout_regularization=self.dropout_regularization,
                         dropout=self.dropout,
                         convinput=conv_good_size,
                         size=self.size_layers[0],
                         stride=(2,2,2),
                          residual_connection=self.residual_connection,
                          number_conv=self.number_conv)
    
        conv2=self.UNet_conv_layers(start_neurons=self.start_neurons,
                         kernel_size=self.kernel_size,
                         maxpool=self.maxpool,
                         dropout_regularization=self.dropout_regularization,
                         dropout=self.dropout,
                         convinput=conv1,
                         size=self.size_layers[1],
                         stride=self.stride,
                          residual_connection=self.residual_connection,
                          number_conv=self.number_conv)
    
        conv3=self.UNet_conv_layers(start_neurons=self.start_neurons,
                         kernel_size=self.kernel_size,
                         maxpool=self.maxpool,
                         dropout_regularization=self.dropout_regularization,
                         dropout=self.dropout,
                         convinput=conv2,
                         size=self.size_layers[2],
                         stride=self.stride,
                          residual_connection=self.residual_connection,
                          number_conv=self.number_conv)
    
        # middel layer 
        conv_middle = self.UNet_conv_layers(start_neurons=self.start_neurons,
                     kernel_size=self.kernel_size,
                     maxpool=self.maxpool,
                     dropout_regularization=self.dropout_regularization,
                     dropout=self.dropout,
                     convinput=conv3,
                     size=self.size_layers[3],
                     stride=self.stride,
                     residual_connection=self.residual_connection,
                     number_conv=self.number_conv)
    
    
    
    
        # decoder -------------------------------------------------
    
    
        uconv3=self.UNet_deconv_layers(start_neurons=self.start_neurons,
                             kernel_size=self.kernel_size,
                             stride=self.stride,
                            dropout_regularization=self.dropout_regularization,
                             dropout=self.dropout,
                             deconvinput=conv_middle,
                             convconcatenate=conv3,
                             size=sizes_down[0],
                             residual_connection=self.residual_connection,
                             number_conv=self.number_conv)   
    
    
    
        uconv2=self.UNet_deconv_layers(start_neurons=self.start_neurons,
                             kernel_size=self.kernel_size,
                             stride=self.stride,
                            dropout_regularization=self.dropout_regularization,
                             dropout=self.dropout,
                             deconvinput=uconv3,
                             convconcatenate=conv2,
                             size=sizes_down[1],
                             residual_connection=self.residual_connection,
                             number_conv=self.number_conv)   
    
        uconv1=self.UNet_deconv_layers(start_neurons=self.start_neurons,
                             kernel_size=self.kernel_size,
                             stride=(2,2,2),
                             dropout_regularization=self.dropout_regularization,
                             dropout=self.dropout,
                             deconvinput=uconv2,
                             convconcatenate=conv1,
                             size=sizes_down[2],
                             residual_connection=self.residual_connection,
                             number_conv=self.number_conv,
                             dilation_rate=2)  
    
        #-----------------------------------------------------------
        self.output_layer = Conv3D(1, kernel_size=self.kernel_size, 
                                   padding="same", 
                                   activation="relu",
                                   strides=self.stride)(uconv1)
    
        #return output_layer
    
    
    # --------------------------------------------------------------------------------------
    def disc_model(self):
        
        # for data size consistency ----------------------------------------
        conv_good_size = ZeroPadding3D(padding=((0, 0), (0, 0), (0, 128-self.size_input[3])))(self.input_layer)
        # --------------------------------------------

        conv1=self.UNet_conv_layers(start_neurons=self.start_neurons,
                         kernel_size=self.kernel_size,
                         maxpool=self.maxpool,
                         dropout_regularization=self.dropout_regularization,
                         dropout=self.dropout,
                         convinput=conv_good_size,
                         size=self.size_layers[0],
                         stride=(2,2,2),
                          residual_connection=self.residual_connection,
                          number_conv=self.number_conv)
    
        conv2=self.UNet_conv_layers(start_neurons=self.start_neurons,
                         kernel_size=self.kernel_size,
                         maxpool=self.maxpool,
                         dropout_regularization=self.dropout_regularization,
                         dropout=self.dropout,
                         convinput=conv1,
                         size=self.size_layers[1],
                         stride=self.stride,
                          residual_connection=self.residual_connection,
                          number_conv=self.number_conv)
    
        conv3=self.UNet_conv_layers(start_neurons=self.start_neurons,
                         kernel_size=self.kernel_size,
                         maxpool=self.maxpool,
                         dropout_regularization=self.dropout_regularization,
                         dropout=self.dropout,
                         convinput=conv2,
                         size=self.size_layers[2],
                         stride=self.stride,
                          residual_connection=self.residual_connection,
                          number_conv=self.number_conv)
    
        # middel layer 
        conv_middle = self.UNet_conv_layers(start_neurons=self.start_neurons,
                     kernel_size=self.kernel_size,
                     maxpool=self.maxpool,
                     dropout_regularization=self.dropout_regularization,
                     dropout=self.dropout,
                     convinput=conv3,
                     size=self.size_layers[3],
                     stride=self.stride,
                     residual_connection=self.residual_connection,
                     number_conv=self.number_conv)
        
        flatten=Flatten()(conv_middle)
        
        self.output_layer = Dense(1,activation='sigmoid')(flatten)
        
        
    
    
 
        
    
    # -------------------------------------------------------------------------------------------------------------------
    def UNet_conv_layers(self,start_neurons,
                         kernel_size,
                         maxpool,
                         dropout_regularization,
                         dropout,
                         convinput,
                         size,stride,
                         number_conv=1,
                         residual_connection=False):
        
        # first layer -------------------------------------------------------------------------------------
        conv_init = Conv3D(start_neurons * size, kernel_size, activation="relu",strides=stride,padding='same',
                           kernel_initializer=self.kernel_initializer,
                           bias_initializer=self.bias_initializer)(convinput)
        
        conv_init=tf.keras.layers.Activation('relu')(conv_init)

        conv=conv_init
        
        
        if self.number_conv>=2:
            for l in np.arange(self.number_conv-2):
                conv = Conv3D(self.start_neurons * self.size_layers, kernel_size, 
                              activation="relu", 
                              padding="same",
                              strides=(1,1,1),
                              kernel_initializer=self.kernel_initializer,
                              bias_initializer=self.bias_initializer)(conv)
                conv=tf.keras.layers.Activation('relu')(conv)
                
                if dropout_regularization:
                    conv = Dropout(dropout)(conv)
            
            
        if residual_connection:
                conv=Add()([conv,conv_init])
            
        # last layer ---------------------------
        conv = Conv3D(start_neurons * size, kernel_size, 
                          activation="relu", 
                          padding="same",
                          strides=(1,1,1),
                          kernel_initializer=self.kernel_initializer,
                          bias_initializer=self.bias_initializer)(conv)
        conv=tf.keras.layers.Activation('relu')(conv)

        
        return conv
  
    #---------------------------------------------


        
        
    
  # -------------------------------------------------------------------------------------------------------------------  
    
    def UNet_deconv_layers(self,start_neurons,kernel_size,stride,dropout_regularization,dropout,deconvinput,convconcatenate,size,number_conv=1,residual_connection=False,dilation_rate=1):
        
        # note : at least 2 deconv layers in this script
        
        uconv = concatenate([deconvinput, convconcatenate]) 
        deconv_init = Conv3DTranspose(start_neurons * size, kernel_size=kernel_size, 
                                      strides=stride,dilation_rate=dilation_rate,padding='same',
                                     kernel_initializer=self.kernel_initializer,
                                       bias_initializer=self.bias_initializer)(uconv)
        
        deconv_init=tf.keras.layers.Activation('relu')(deconv_init)
        
        #--------------------------------------------------------------
        deconv = deconv_init
        #---------------------------------------------------------------
                
        if number_conv>=2:
            for l in np.arange(number_conv-2):
                deconv = Conv3DTranspose(start_neurons * size, 
                                         kernel_size, 
                                         strides=(1,1,1), 
                                         padding="same",
                                         kernel_initializer=self.kernel_initializer,
                                           bias_initializer=self.bias_initializer)(deconv)
                
                deconv=tf.keras.layers.Activation('relu')(deconv)
                
                if dropout_regularization:
                           deconv = Dropout(self.dropout)(deconv)
                
        if residual_connection:
                deconv=Add()([deconv,deconv_init])
            
        
        # last convolutional layer -------------------
        deconv = Conv3DTranspose(start_neurons * size, 
                                        kernel_size, 
                                        strides=(1,1,1), 
                                        padding="same",
                                        kernel_initializer=self.kernel_initializer,
                                       bias_initializer=self.bias_initializer)(deconv)  
        
        deconv=tf.keras.layers.Activation('relu')(deconv)
        
        return deconv
    
    
    
    # ----------------------------------------------------------------------------
    def callmodel(self):
        model = Model(inputs=self.input_layer,
                  outputs=self.output_layer,
                  name=self.model_name)
        return model
```

```python
i=0
def train_gan_model(data_train_batch,data_test_batch,config):
    
    g_loss_best=1e10
    tf.keras.backend.clear_session()
    for epochs in range(config.epoch):
    
    
        for (X,y) in data_train_batch:
        
            
            batch_size=X.shape[0]
            # generate images ------------------------------------------
            fake_x = gen.predict(X)       
            real_x=y
            x=np.concatenate((real_x,fake_x))
                
            
            # discriminator -----------------------------------
            # half of the disc y to 0.9 to correlate with x       
            disc_y = np.zeros(2*batch_size)
            disc_y[:batch_size] = 0.9
        
            # train discriminator -----------------------------
            d_loss = disc.train_on_batch(x,disc_y)
            GANlosses(gen_model=gen,x=X, y=y, disc_loss=d_loss)
            
            # train GAN -----------------------------------------
            y_gen = np.ones(batch_size)
            g_loss = gan.train_on_batch(X,y_gen)
            
            # check gen loss ----------------------------------
            #gen_loss = np.sum(np.power((real_x-fake_x),2))/batch_size
        
        
        print('epoch ' + str(epochs) + '/' + str(config.epoch) )
    
        # save current model
        track=track_model(disc,gen,gan,output_dir+str(i))
        track.calculate_total_loss(data_test_batch)
        #track.save_best_model()
        if g_loss_best>g_loss:
            track.save_other_models(track.model_gen,track.best_model_name_gen)
            track.save_other_models(track.model_gan,track.best_model_name_gan)
            track.save_other_models(track.model_disc,track.best_model_name_disc)
            g_loss_best=g_loss
        
    
    
        ## TO DO -------------------------------------------------------------------------------------------
     
        # store losses in wandb ----------------
        # store:
        # epoch, g loss, d_loss, val loss
        wandb.log({"train_loss_discriminator":d_loss,
                  "train_loss_gan":g_loss,
                   "train_loss_gen":gen_loss,
                   "val_loss_gen":track.validation_loss_gen_current,
                   "val_loss_disc":track.validation_loss_disc_current,
                   "val_loss_gan":track.validation_loss_total_current,
                   "val_loss_total_best":track.validation_loss_total_best
                  })
    
    
    
    wandb.run.finish()
```

```python
class track_model:
    def __init__(self,
                 model_disc,
                 model_gen,
                 model_gan,
                 save_path_model):
        
        
        self.save_path_model = save_path_model
        self.check_dir(save_path_model)
        
        # disc -----------------------------------------------------
        self.model_name_disc='disc'
        self.best_model_name_disc=save_path_model+'/'+self.model_name_disc+'_best_model'
        self.model_disc=model_disc
        
        # gen -----------------------------------------------------
        self.model_name_gen='gen'
        self.best_model_name_gen=save_path_model+'/'+self.model_name_gen+'_best_model'
        self.model_gen=model_gen
        
        # gen -----------------------------------------------------
        self.model_name_gan='gan'
        self.best_model_name_gan=save_path_model+'/'+self.model_name_gan+'_best_model'
        self.model_gan=model_gan
        
        # tracked loss ------------------------------------------
        self.validation_loss_disc_best=1e10
        self.validation_loss_disc_current=1e10
        self.validation_loss_gen_current = 1e10
        
        self.validation_loss_total_current = 1e10
        self.validation_loss_total_best = 1e10
    
    # ------------------------------------------------------------------------
    def save_optimizer_weights(self,model,model_name):
        # save optimizer weights ------------------------------
        #model.save_weights('weights.h5')
        symbolic_weights = getattr(model.optimizer, 'weights')
        weight_values = tf.keras.backend.batch_get_value(symbolic_weights)
        with open(self.save_path_model+'/optimizer_'+ model_name+'.pkl', 'wb') as f:
                pickle.dump(weight_values, f)
        # how to load the optimizer weights:
        #model._make_train_function()
        # How to set weights to optimizer model
        #with open('optimizer.pkl', 'rb') as f:
        #    weight_values = pickle.load(f)
        #model.optimizer.set_weights(weight_values)
        

    # -------------------------------------------------------------------------    
    #def save_current_model(self):
    #    if not os.path.dirname(self.best_model_name):
    #        self.model.save(self.best_model_name)
            
     #   else:
     #       self.model.save(save_path_model+'/current_model')
    # -----------------------------------------------------------------------        
    def save_other_models(self,model,model_name):
        model.save(model_name)
        
    # ------------------------------------------------------------------------    
    def save_best_model(self):
        
        print('validation total = ' +str(self.validation_loss_total_current) )
            
        if self.validation_loss_total_current < self.validation_loss_total_best:
            # save gen
            self.model_gen.save(self.best_model_name_gen)
            self.validation_loss_gen_best=self.validation_loss_gen_current
            #self.save_optimizer_weights(self.model_gen,'gen')
            
            # save disc
            self.model_disc.save(self.best_model_name_disc)
            self.validation_loss_disc_best=self.validation_loss_disc_current
            self.save_optimizer_weights(self.model_disc,'disc')
            
            # save GAN
            self.model_disc.save(self.best_model_name_gan)
            self.validation_loss_total_best=self.validation_loss_total_current
            self.save_optimizer_weights(self.model_gan,'gan')
            
            print('save_models : ' + str(self.best_model_name_gen))
            
    #-------------------------------------------------------------------------    
    #def calculate_validation_loss_gen(self,data_test_batch):
        
        
     #   for (X,y) in data_test_batch:
     #       current_images=self.model.predict(X)
            
            
     #       validation_loss=np.sum(np.power(y-current_images,2))/len(X)
     #       val_loss = val_loss+validation_loss
        
     #  self.validation_loss_gen_current=val_loss/len(data_test_batch)
        
        
    #----------------------------------------------------------------------- 
    #def calculate_validation_loss_disc(self,data_test_batch):
        
     #   for (X,y) in data_test_batch:
     #       fake_im=self.model_gen.predict(X)
     #       #validation_loss_data=sum(np.power(y-fake_im,2))/len(X)
            
     #       disc_pred= self.model_disc.predict(fake_im)
     #       bce= tf.keras.losses.binary_crossentropy()
     #       validation_loss_disc = bce(y,disc_pred)
            
     #       total_loss=validation_loss_data+validation_loss_disc
            
            
     #       val_loss = val_loss+total_loss
        
     #   self.validation_loss_disc_current=val_loss/len(data_test_batch)
    # ------------------------------------------------------------------------    
    def calculate_total_loss(self,data_test_batch):
        
        val_loss_gen=0
        val_loss_disc=0
        #loss_gen = tf.keras.losses.mse
        loss_disc = tf.keras.losses.BinaryCrossentropy()
        
        for (X,y) in data_test_batch:
            
            batch_size=X.shape[0]
            
            # gen
            fake_im=self.model_gen.predict(X)
            
            validation_loss_gen = np.sum(np.power((y-fake_im),2))/batch_size
            #validation_loss_gen=self.model_gen.evaluate(y,fake_im)
            
            val_loss_gen = val_loss_gen+validation_loss_gen
            
                      
            # disc
            im=np.concatenate((y,fake_im))           
            disc_y = np.zeros(2*batch_size)
            disc_y[:batch_size] = 0.9
            disc_pred= self.model_disc.predict(im)
            
            
            validation_loss_disc = self.model_disc.evaluate(disc_y,disc_pred)#loss_disc(disc_y,disc_pred).numpy()
            val_loss_disc = val_loss_disc + validation_loss_disc
            
        self.validation_loss_gen_current = val_loss_gen/len(data_test_batch)    
        self.validation_loss_disc_current = val_loss_disc/len(data_test_batch)    
        self.validation_loss_total_current=(val_loss_gen+val_loss_disc)/len(data_test_batch)
            
            


# ---------------------------------------------------------------------------

    def check_dir(self,name_dir):
        if not os.path.dirname(name_dir):
            os.makedirs(name_dir)
    
```
